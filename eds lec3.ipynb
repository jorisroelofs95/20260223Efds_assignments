## user defined functions

```python
from math import log, exp
def myMean( x, isGeom = False ):
 """
 Calculate the arithmetic or geometric mean of X
 """
 if len(x) == 0:
  raise RuntimeError("This funtion requires a nonempty list")
 if isGeom: 
  #calc log of all numbers
  logofx = [log(x=n) for n in x]
  #calc mean of all logs
  meanLogofX = sum(logofx) / len(x)
  return exp(meanLogofX)
 else:
  return sum(x)/len(x)
  
  
```

```python
nums = [1,2,3,4]
myMean( x=nums )
```

```python
nums = []
myMean( x=nums )
```

```python
course2grades = {
    "math": (7,6,8,9),
    "physics": (7,7,6,5)
}
{C:myMean(x=gs) for C, gs in course2grades.items()}
```
